# Model Experimentation 

```python
import pandas as pd

def pl(data, model, predictors, start=4, step=1):
    all_predictions = []
    
    seasons = sorted(data["season"].unique())
    
    for i in range(start, len(seasons), step):
        season = seasons[i]
        train = data[data["season"] < season]
        test = data[data["season"] == season]
        
        # Perform one-hot encoding for the "season" column
        train = pd.get_dummies(train, columns=["season"], drop_first=True)
        test = pd.get_dummies(test, columns=["season"], drop_first=True)
        
        # Ensure all predictors are numeric
        train = train.astype({'is_top4_next_season': 'int'})  # Ensure target variable is integer
        test = test.astype({'is_top4_next_season': 'int'})    # Ensure target variable is integer
        
        model.fit(train[predictors], train["is_top4_next_season"])
        
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["is_top4_next_season"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        
        all_predictions.append(combined)
    
    return pd.concat(all_predictions)
```

```python
predictors = [col for col in data_encoded.columns if col not in ["is_top4_next_season", "season"]]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score


# List of models to test
models = [
    LogisticRegression(),
    RandomForestClassifier(),
    XGBClassifier(),
    SVC(),
    KNeighborsClassifier(),
    GaussianNB()
]

# Iterate through the models and call the 'pl' function
for model in models:
    print(f"Model: {model.__class__.__name__}")
    result = pl(data_encoded, model, predictors, start=4, step=1)
    # You can add code here to analyze the results or print evaluation metrics
    print(result)  # Print the combined results for each model
    print("accuracy score :", accuracy_score(result["actual"], result["prediction"]))
    print("\n")
```

# Model Evaluation 

```python
# we will use TimeSeriesSplit instead of cross validation because we will use only the previous data 
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import accuracy_score

# Define the number of splits (you can adjust this)
n_splits = 10
tscv = TimeSeriesSplit(n_splits=n_splits)

# Initialize a list to store evaluation results
evaluation_results = []

# Iterate through the splits
for train_index, test_index in tscv.split(data_encoded):
    train_data = data_encoded.iloc[train_index]
    test_data = data_encoded.iloc[test_index]
    
    # Split predictors and target variables
    train_X = train_data[predictors]
    train_y = train_data["is_top4_next_season"]
    test_X = test_data[predictors]
    test_y = test_data["is_top4_next_season"]
    
    # Train your model (replace 'model' with your actual model)
    model.fit(train_X, train_y)
    
    # Make predictions
    predictions = model.predict(test_X)
    
    # Evaluate the model using the accuracy score (you can use other metrics)
    accuracy = accuracy_score(test_y, predictions)
    
    # Store the evaluation result for this split
    evaluation_results.append(accuracy)

# Calculate the mean and standard deviation of the evaluation results
mean_accuracy = sum(evaluation_results) / n_splits
std_accuracy = np.std(evaluation_results)

print(f"Mean Accuracy: {mean_accuracy}")
print(f"Standard Deviation: {std_accuracy}")
```

# Model Tuned Performance
After getting the baselines, let's see if we can improve on the indivdual model results!I mainly used grid search to tune the models. I also used Randomized Search for the Random Forest and XG boosted model to simplify testing time.

```python
# Create the target variable
train_y = data_encoded['is_top4_next_season']
```

```python
# Drop specific columns from your DataFrame
#columns_to_exclude = ['is_top4_next_season']
#data_encoded.drop(columns=columns_to_exclude, inplace=True)

# Create train_X with the remaining columns
train_X = data_encoded[predictors]
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# Define your models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "XGBoost": XGBClassifier(),
    "SVC": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Gaussian Naive Bayes": GaussianNB()
}

# Define hyperparameter grids for each model
param_grids = {
    "Logistic Regression": {
        'max_iter': [2000],
        'penalty': ['l1', 'l2'],
        'C': [0.001, 0.01, 0.1, 1, 10],
        'solver': ['liblinear']
    },
    "Random Forest": {
        'n_estimators': [10, 50, 100, 200],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10]
    },
    "XGBoost": {
        'n_estimators': [100, 200, 300],
        'max_depth': [3, 4, 5],
        'learning_rate': [0.01, 0.1, 0.2]
    },
    "SVC": {
        'C': [0.01, 0.1, 1, 10],
        'kernel': ['linear', 'rbf', 'poly']
    },
    "K-Nearest Neighbors": {
        'n_neighbors': [3, 5, 7, 9],
        'p': [1, 2]
    },
    "Gaussian Naive Bayes": {
        'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6]
    }
}

# Perform time series cross-validation for each model
best_models = {}

# Define the number of splits
n_splits = 10
tscv = TimeSeriesSplit(n_splits=n_splits)

for model_name, model in models.items():
    print(f"Tuning {model_name} using Time Series Cross-Validation...")
    param_grid = param_grids[model_name]
    
    # Initialize a list to store evaluation results
    evaluation_results = []
    
    # Iterate through the splits
    for train_index, test_index in tscv.split(train_X):
        train_X_split, test_X_split = train_X.iloc[train_index], train_X.iloc[test_index]
        train_y_split, test_y_split = train_y.iloc[train_index], train_y.iloc[test_index]
        
        grid_search = GridSearchCV(model, param_grid=param_grid, cv=5, verbose=2, n_jobs=-1)
        grid_search.fit(train_X_split, train_y_split)
        
        best_models[model_name] = grid_search
        
        # Store the evaluation result for this split
        evaluation_results.append(grid_search.best_score_)
    
    # Print the best parameters and the mean cross-validation score
    print(f"Best parameters for {model_name}: {best_models[model_name].best_params_}")
    mean_cv_score = sum(evaluation_results) / n_splits
    print(f"Mean cross-validation score for {model_name}: {mean_cv_score}")
```
